```python
def add_days_before_failure_col_in_failures_merged(row): 
    days_diff = (datetime.strptime(row['date_y'], '%Y-%m-%d') - datetime.strptime(row['date_x'], '%Y-%m-%d'))
    days = int(divmod(days_diff.total_seconds(), 86400)[0])
    row['days_before_failure'] = days
    return row

failures = data[data['failure'] == 1]
display(failures.head())
failures_merged = data.merge(failures, on=['serial_number', 'model'], right_index=True)
failures_merged_final = failures_merged.apply(add_days_before_failure_col_in_failures_merged, axis = 1)
failures_merged_final.head()
```
